```python
def Dic(feat_l_tr,feat_r_tr,a): #入力の配列は2次元配列である必要がある
    feat_tr = np.concatenate([feat_l_tr, feat_r_tr],a)
    return feat_tr
```

```python
def SVM(filename1,filename2,filename3,filename4):
    #対象被験者の特徴ベクトルをMATLABから入手
    feat_C1 = np.loadtxt(filename1,delimiter=",")
    feat_C2 = np.loadtxt(filename2,delimiter=",")
    feat_Test_C1 = np.loadtxt(filename3,delimiter=",")
    feat_Test_C2 = np.loadtxt(filename4,delimiter=",")
    label_left = np.ones(feat_C1.shape[0])
    label_right = np.full(feat_C1.shape[0],2)
    feat = Dic(feat_C1,feat_C2,0)
    label = Dic(label_left,label_right,0)
    from sklearn.svm import SVC
    model = LinearSVC()
    model.fit(feat,label)
    answer1 = model.predict(feat_Test_C1)
    answer2 = model.predict(feat_Test_C2)
    x1 = 0
    x2 = 0
    for i in range(answer1.shape[0]):
        if answer1[i] == 1:
            x1 = x1 + 1
        if answer2[i] == 2:
            x2 = x2 + 1
    temp1 = ((x1)/(answer1.shape[0]))*100
    temp2 = ((x2)/(answer2.shape[0]))*100
    temp = ((x1+x2)/(answer1.shape[0]+answer2.shape[0]))*100

    return temp,temp1,temp2
```
